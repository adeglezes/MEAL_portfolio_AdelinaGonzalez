# Simulated Indicator Trend

This notebook demonstrates a simple simulated monitoring indicator over four quarters.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(123)

# Simulated indicator across 4 quarters
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
values = np.cumsum(np.random.normal(5, 2, 4)) + 50

df = pd.DataFrame({'Quarter': quarters, 'IndicatorValue': values})
df
```

```python
plt.plot(df['Quarter'], df['IndicatorValue'], marker='o')
plt.title('Simulated Indicator Trend')
plt.ylabel('Indicator Value')
plt.xlabel('Quarter')
plt.grid(True)
plt.show()
```
